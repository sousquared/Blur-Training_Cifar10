```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# training loss

```python
# load data
normal = pd.read_csv('data/run-seed42_normal_b64-tag-loss_train.csv')['Value'].values
steps = pd.read_csv('data/run-step_e50-tag-loss_train.csv')['Value'].values

# plot
x = range(1, 51)  # x range
fig, ax = plt.subplots()
plt.title("training loss")
ax.plot(x, normal, label='normal')
ax.plot(x, steps, label='step-by-step')
ax.legend()
ax.set_xlabel('Epoch')
ax.set_ylabel('CrossEntropyLoss')
ax.text(2, 2.1, 'σ=3 k(9,9)')
ax.text(10.5, 1.1, 'σ=2 k(7,7)')
ax.text(20.5, 0.6, 'σ=1.5 k(5,5)')
ax.text(30.5, 0.35, 'σ=1 k(3,3)')
ax.text(40.5, 0.3, 'raw')
ax.grid(True)
fig.show()
fig.savefig('blur-steps_training.png')
```

# test loss

```python
normal = pd.read_csv('data/run-seed42_normal_b64-tag-loss_val.csv')['Value'].values
steps = pd.read_csv('data/run-step_e50-tag-loss_val.csv')['Value'].values

# plot
x = range(1, 51)  # x range
fig, ax = plt.subplots()
plt.title("test loss")
ax.plot(x, normal, label='normal')
ax.plot(x, steps, label='step-by-step')
ax.legend()
ax.set_xlabel('Epoch')
ax.set_ylabel('CrossEntropyLoss')
ax.text(3, 2.04, 'σ=3 k(9,9)')
ax.text(10.5, 1.9, 'σ=2 k(7,7)')
ax.text(20.7, 1.7, 'σ=1.5 k(5,5)')
ax.text(30.5, 1.45, 'σ=1 k(3,3)')
ax.text(40.5, 1.25, 'raw')
ax.grid(True)
fig.show()
fig.savefig('blur-steps_test.png')
```
