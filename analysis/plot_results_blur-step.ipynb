# plot_results_blur-step

```python
%matplotlib inline
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
RESULT_DIR = './analysis_results/'
```

# training loss

```python
# load data
normal = pd.read_csv('data/run-normal_60e-tag-loss_train.csv')['Value'].values
steps = pd.read_csv('data/run-blur-step_s4-3-2-1_150e-tag-loss_train.csv')['Value'].values[:60]

# plot
x = range(1, 61)  # x range
fig, ax = plt.subplots()
plt.title("training loss")
ax.plot(x, normal, label='normal')
ax.plot(x, steps, label='step-by-step')
ax.legend()
ax.set_xlabel('Epoch')
ax.set_ylabel('CrossEntropyLoss')
ax.text(2, 2.1, 'σ=4')
ax.text(10.5, 1.2, 'σ=3')
ax.text(20.5, 0.8, 'σ=2')
ax.text(30.5, 0.55, 'σ=1')
ax.text(40.5, 0.4, 'raw')
ax.grid(True)
fig.show()
fig.savefig(RESULT_DIR + 'blur-steps_loss_train.png')
```

# validation loss

```python
normal = pd.read_csv('data/run-normal_60e-tag-loss_val.csv')['Value'].values
steps = pd.read_csv('data/run-blur-step_s4-3-2-1_150e-tag-loss_val.csv')['Value'].values[:60]

# plot
x = range(1, 61)  # x range
fig, ax = plt.subplots()
plt.title("test loss")
ax.plot(x, normal, label='normal')
ax.plot(x, steps, label='step-by-step')
ax.legend()
ax.set_xlabel('Epoch')
ax.set_ylabel('CrossEntropyLoss')
ax.text(1, 3.1, 'σ=4')
ax.text(11, 3.1, 'σ=3')
ax.text(21, 2.2, 'σ=2')
ax.text(31, 1.35, 'σ=1')
ax.text(41, 1.2, 'raw')
ax.grid(True)
fig.show()
fig.savefig(RESULT_DIR + 'blur-steps_loss_val.png')
```
