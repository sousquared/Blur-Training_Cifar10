```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
RESULT_DIR = './results/blur/'
```

# training loss

```python
normal = pd.read_csv('./data/run-normal_60e-tag-loss_train.csv')['Value'].values[:60]
s1 = pd.read_csv('./data/run-blur-half-data_s1_k7-7_90e-tag-loss_train.csv')['Value'].values[:60]
s2 = pd.read_csv('./data/run-blur-half-data_s2_k13-13_90e-tag-loss_train.csv')['Value'].values[:60]
s3 = pd.read_csv('./data/run-blur-half-data_s3_k19-19_90e-tag-loss_train.csv')['Value'].values[:60]
s4 = pd.read_csv('./data/run-blur-half-data_s4_k25-25_90e-tag-loss_train.csv')['Value'].values[:60]

# plot
x = range(1, 61)  # x range
fig, ax = plt.subplots()
plt.title("training loss")
ax.plot(x, normal, label='normal')
ax.plot(x, s1, label='σ=1')
ax.plot(x, s2, label='σ=2')
ax.plot(x, s3, label='σ=3')
ax.plot(x, s4, label='σ=4')
ax.legend()
ax.set_xlabel('Epoch')
ax.set_ylabel('CrossEntropyLoss')
ax.grid(True)
fig.show()
fig.savefig(RESULT_DIR + 'blur-half-data_loss_train.png')
```
