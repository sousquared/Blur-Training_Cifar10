```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# training loss

```python
normal = pd.read_csv('data/run-seed42_normal_b64-tag-loss_train.csv')['Value'].values
k3 = pd.read_csv('data/run-s1_k3-3_noblur_from25-tag-loss_train.csv')['Value'].values
k5 = pd.read_csv('data/run-s1.5_k5-5_noblur_from25-tag-loss_train.csv')['Value'].values
k7 = pd.read_csv('data/run-s2_k7-7_noblur_from25-tag-loss_train.csv')['Value'].values
k9 = pd.read_csv('data/run-s3_k9-9_noblur_from25-tag-loss_train.csv')['Value'].values

# plot
x = range(1, 51)  # x range
fig, ax = plt.subplots()
plt.title("training loss")
ax.plot(x, normal, label='normal')
ax.plot(x, k3, label='σ=1, kernel=(3,3)')
ax.plot(x, k5, label='σ=1.5, kernel=(5,5)')
ax.plot(x, k7, label='σ=2, kernel=(7,7)')
ax.plot(x, k9, label='σ=3, kernel=(9,9)')
ax.legend()
ax.set_xlabel('Epoch')
ax.set_ylabel('CrossEntropyLoss')
ax.grid(True)
fig.show()
fig.savefig('blur-half_training.png')
```

# test loss

```python
normal = pd.read_csv('data/run-seed42_normal_b64-tag-loss_val.csv')['Value'].values
k3 = pd.read_csv('data/run-s1_k3-3_noblur_from25-tag-loss_val.csv')['Value'].values
k5 = pd.read_csv('data/run-s1.5_k5-5_noblur_from25-tag-loss_val.csv')['Value'].values
k7 = pd.read_csv('data/run-s2_k7-7_noblur_from25-tag-loss_val.csv')['Value'].values
k9 = pd.read_csv('data/run-s3_k9-9_noblur_from25-tag-loss_val.csv')['Value'].values

# plot
x = range(1, 51)  # x range
fig, ax = plt.subplots()
plt.title("test loss")
ax.plot(x, normal, label='normal')
ax.plot(x, k3, label='σ=1, kernel=(3,3)')
ax.plot(x, k5, label='σ=1.5, kernel=(5,5)')
ax.plot(x, k7, label='σ=2, kernel=(7,7)')
ax.plot(x, k9, label='σ=3, kernel=(9,9)')
ax.legend()
ax.set_xlabel('Epoch')
ax.set_ylabel('CrossEntropyLoss')
ax.grid(True)
fig.show()
fig.savefig('blur-half_test.png')
```
